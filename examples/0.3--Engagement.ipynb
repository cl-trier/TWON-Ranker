```python
import datetime
import typing

from src import modules
```

```python
MINIMUM: float = 0.2
REFERENCE_TIMEDELTA: datetime.timedelta = datetime.timedelta(days=3)

N: int = 1_000
REFERENCE_DATETIME: datetime.datetime = datetime.datetime.now()
```

```python
DECAY = modules.Decay(minimum=MINIMUM, reference_timedelta=REFERENCE_TIMEDELTA)
```

```python
observations: typing.List[datetime.datetime] = [
    REFERENCE_DATETIME - (REFERENCE_TIMEDELTA * i / N)
    for i in reversed(range(N))
]
observations
```

```python
print(f'''engagement for observations ({N=}):

{'count_based:':20} {modules.Engagement(func='count_based', log_normalize=False)(items=observations)} 
{'count_based (log):':20} {modules.Engagement(func='count_based',log_normalize=True)(items=observations)} 

{'decay_based:':20}  {modules.Engagement(func='decay_based', log_normalize=False)(items=observations, reference_datetime=REFERENCE_DATETIME, decay=DECAY)}
{'decay_based (log):':20}  {modules.Engagement(func='decay_based', log_normalize=True)(items=observations, reference_datetime=REFERENCE_DATETIME, decay=DECAY)}
''')
```
