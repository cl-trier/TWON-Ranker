```python
import datetime

MINIMUM: float = 0.2
REFERENCE_TIMEDELTA: datetime.timedelta = datetime.timedelta(days=3)

N: int = 1_000
REFERENCE_DATETIME: datetime.datetime = datetime.datetime.now()
```

```python
import typing

observations: typing.List[datetime.datetime] = [
    REFERENCE_DATETIME - (REFERENCE_TIMEDELTA * i / N) 
    for i in reversed(range(N))
]
observations
```

```python
from src import modules

FUNCTION = modules.Decay(minimum=MINIMUM, reference_timedelta=REFERENCE_TIMEDELTA)
```

```python
import seaborn as sns

sns.lineplot(
    x = observations,
    y = [FUNCTION(obs, REFERENCE_DATETIME) for obs in observations],
)
```
